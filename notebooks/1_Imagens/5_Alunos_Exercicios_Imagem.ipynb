**4) Desenhe um TRIÂNGULO AZUL no meio da imagem. O tamanho e o ângulo ficam a seu critério, mas deve ser um triângulo (três lados) em qualquer orientação.**

```python
draw_img = img_rgb.copy()

pt1 = (450, 350)
pt2 = (200, 700)
pt3 = (700, 700)
color = (0, 0, 255)
thickness = 10

cv2.line(draw_img, pt1, pt2, color=color, thickness=thickness)
cv2.line(draw_img, pt2, pt3, color=color, thickness=thickness)
cv2.line(draw_img, pt3, pt1, color=color, thickness=thickness)

# finding centroid using the following formula 
# (X, Y) = (x1 + x2 + x3//3, y1 + y2 + y3//3)
centroid = ((pt1[0]+pt2[0]+pt3[0])//3, (pt1[1]+pt2[1]+pt3[1])//3)

# Drawing the centroid on the window  
final = cv2.circle(draw_img, centroid, 1, (0, 255, 0)) 

plt.imshow(final)
```

```python
# Exercício 4 - solução 2:

image = img_rgb.copy()

pt1 = (450, 350)
pt2 = (200, 700)
pt3 = (700, 700)

color = (0, 0, 255)
thickness = 10

cv2.circle(image, pt1, 2, (0, 255, 0) ,-1)
cv2.circle(image, pt2, 2, (0, 255, 0) ,-1)
cv2.circle(image, pt3, 2, (0, 255, 0) ,-1)

triangle = np.array( [pt1, pt2, pt3] )

cv2.drawContours(image, [triangle], 0, color=color, thickness=thickness)

plt.imshow(image)
```
