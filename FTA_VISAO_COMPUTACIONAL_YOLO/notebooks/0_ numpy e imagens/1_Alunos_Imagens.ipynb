## **Separando as camadas R, G e B**

```python
# mostra a imagem colorida ../DATA/rgb_image.jpg

img = cv2.imread("../DATA/3d_array.png")
display(img)
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) # converte RGB em GBR
plt.imshow(img)
```

```python
# mostra a camada vermelha

img = cv2.imread("../DATA/3d_array.png")
display(img[:,:,0])
img_red = img[:,:,0]
plt.imshow(img_red, cmap='Reds')
```

```python
# mostra a camada verde

img = cv2.imread("../DATA/3d_array.png")
display(img[:,:,1])
img_red = img[:,:,1]
plt.imshow(img_red, cmap='Greens')
```

```python
# mostra a camada azul

img = cv2.imread("../DATA/3d_array.png")
display(img[:,:,2])
img_red = img[:,:,2]
plt.imshow(img_red, cmap='Blues')
```

## **Como  a decomposio das camadas numa foto**

```python
# mostra a foto do quafit ../DATA/quafit.jpeg

img = cv2.imread("../DATA/quafit.jpeg")
# img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) # converte RGB em GBR
plt.imshow(img)
```

 **Note que usamos a funo COLOR_BGR2RGB. Por padro o OpenCV carrega as imagens na ordem trocada: Ao invs de carregar as camadas de cores na sequencia RGB (vermelho-verde-azul), ele carrega na ordem BGR (azul-verde-vermelho).**  

 **Como vamos mostrar as imagens utilizando o matplotlib  necessrio inverter a ordem das camadas. Caso no faamos isso, as cores ficam distorcidas, vejam:**

```python
pic = cv2.imread('../DATA/quafit.jpeg')
pic = np.asarray(pic)
plt.imshow(pic)
```

```python
# mostra a camada vermelha (zerando as demais camadas)

img = cv2.imread("../DATA/quafit.jpeg")
img_red = img[:,:,0]
plt.imshow(img_red, cmap='Reds')
```

```python
# mostra a camada verde (zerando as demais camadas)

img = cv2.imread("../DATA/quafit.jpeg")
img_red = img[:,:,1]
plt.imshow(img_red, cmap='Greens')
```

```python
# mostra a camada azul (zerando as demais camadas)

img = cv2.imread("../DATA/quafit.jpeg")
img_red = img[:,:,2]
plt.imshow(img_red, cmap='Blues')
```
