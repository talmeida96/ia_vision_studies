```python
# Bibliotecas

import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

# **Desenho em imagens**

## **Criando uma imagem vazia**


**Vamos criar uma imagem vazia com:**
- **512 p√≠xels de altura**
- **512 p√≠xels de largura** 
- **3 camadas**

```python
# Crie uma imagem vazia, de 1 camada
# isto √©, uma matriz de zeros com tamanho 512x512x3

img_black = np.zeros((512, 512, 3))
```

```python
# Mostre as dimens√µes da imagem criada (shape)

img_black.shape
```

```python
# Exiba a imagem na tela usando o Matplotlib

plt.imshow(img_black)
```

---
## **Desenhando formas geom√©tricas**

### üëâ **rectangle()**

**No exemplo de c√≥digo abaixo, um ret√¢ngulo ser√° desenhado na imagem. Para isso s√£o fornecidos os seguintes par√¢metros:**

* **pt_1**: V√©rtice do ret√¢ngulo; (canto superior esquerdo)
* **pt_2**: V√©rtice do ret√¢ngulo oposto a pt1; (canto inferior direito) 
* **color**: Cor do ret√¢ngulo (ou brilho se a imagem for em preto e branco).  
* **thickness**: Espessura das linhas que comp√µem o ret√¢ngulo. Valores negativos, significam que a fun√ß√£o deve desenhar um ret√¢ngulo preenchido.  

```python
# Desenhe um ret√¢ngulo verde no canto superior direito da imagem vazia
# Mostre o resultado 

draw_img = img_black.copy()

start_point = (200, 200) 
end_point = (300, 400) 
color = (255, 0, 0) 
thickness = 2
  
rec_img = cv2.rectangle(draw_img, start_point, end_point, color, thickness)

plt.imshow(rec_img)
```

---
### üëâ **circle()**

**No exemplo de c√≥digo abaixo, um circulo √© adicionado √† imagem utilizando os par√¢metros:**

* **center**: Centro do c√≠rculo
* **radius**: raio do c√≠rculo
* **color**: Cor do c√≠rculo ou brilho (imagem em tons de cinza).  
* **thickness**: Espessura das linhas que comp√µem o c√≠rculo. 

```python
# Desenhe um circulo vermelho no canto superior esquerdo da imagem
# Mostre o resultado 

draw_img = img_black.copy()

center = (250, 250) 
radius = 60
color = (255, 0, 0) 
thickness = 2
  
circle_img = cv2.circle(draw_img, center=center, radius=radius, color=color, thickness=thickness)

plt.imshow(circle_img)
```

---
### üëâ **Figuras preenchidas**

**Para criar uma figura preenchida √© necess√°rio indicar um valor negativo como valor do par√¢metro *tickness***

```python
# Desenhe um circulo vermelho preenchido no canto inferior direito da imagem
# Mostre o resultado 

draw_img = img_black.copy()

center = (250, 250) 
radius = 60
color = (255, 0, 255) 
thickness = -1 # circulo preenchido
  
circle_img = cv2.circle(draw_img, center=center, radius=radius, color=color, thickness=thickness)

plt.imshow(circle_img)
```

---
### üëâ **line()**

```python
# Desenhe uma linha azul diagonal com espessura de 5 p√≠xels

draw_img = img_black.copy()

pt1 = (400, 100) 
pt2 = (100, 400) 
color = (255, 0, 255) 
thickness = 5

line_img = cv2.line(draw_img, pt1=pt1, pt2=pt2, color=color, thickness=thickness)

plt.imshow(line_img)
```

---
### üëâ **putText()**

**Inserindo texto na imagem**

```python
# insira um texto na imagem utilizando a fonte cv2.FONT_HERSHEY_SIMPLEX

draw_img = img_black.copy()

text = "Thayna"
size = 2
fontFace = cv2.FONT_HERSHEY_SIMPLEX
origin = (100, 100)
color = (255, 0, 255) 
thickness = 5

txt_img = cv2.putText(draw_img, text=text, org=origin, fontFace=fontFace, fontScale=size, color=color, thickness=thickness)

plt.imshow(txt_img)
```
