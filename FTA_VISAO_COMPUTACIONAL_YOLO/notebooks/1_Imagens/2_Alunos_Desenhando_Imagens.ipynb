```python
# Bibliotecas

import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

# **Desenho em imagens**

## **Criando uma imagem vazia**


**Vamos criar uma imagem vazia com:**
- **512 píxels de altura**
- **512 píxels de largura** 
- **3 camadas**

```python
# Crie uma imagem vazia, de 1 camada
# isto é, uma matriz de zeros com tamanho 512x512x3

img_black = np.zeros((512, 512, 3))
```

```python
# Mostre as dimensões da imagem criada (shape)

img_black.shape
```

```python
# Exiba a imagem na tela usando o Matplotlib

plt.imshow(img_black)
```

---
## **Desenhando formas geométricas**

### 👉 **rectangle()**

**No exemplo de código abaixo, um retângulo será desenhado na imagem. Para isso são fornecidos os seguintes parâmetros:**

* **pt_1**: Vértice do retângulo; (canto superior esquerdo)
* **pt_2**: Vértice do retângulo oposto a pt1; (canto inferior direito) 
* **color**: Cor do retângulo (ou brilho se a imagem for em preto e branco).  
* **thickness**: Espessura das linhas que compõem o retângulo. Valores negativos, significam que a função deve desenhar um retângulo preenchido.  

```python
# Desenhe um retângulo verde no canto superior direito da imagem vazia
# Mostre o resultado 

draw_img = img_black.copy()

start_point = (200, 200) 
end_point = (300, 400) 
color = (255, 0, 0) 
thickness = 2
  
rec_img = cv2.rectangle(draw_img, start_point, end_point, color, thickness)

plt.imshow(rec_img)
```

---
### 👉 **circle()**

**No exemplo de código abaixo, um circulo é adicionado à imagem utilizando os parâmetros:**

* **center**: Centro do círculo
* **radius**: raio do círculo
* **color**: Cor do círculo ou brilho (imagem em tons de cinza).  
* **thickness**: Espessura das linhas que compõem o círculo. 

```python
# Desenhe um circulo vermelho no canto superior esquerdo da imagem
# Mostre o resultado 

draw_img = img_black.copy()

center = (250, 250) 
radius = 60
color = (255, 0, 0) 
thickness = 2
  
circle_img = cv2.circle(draw_img, center=center, radius=radius, color=color, thickness=thickness)

plt.imshow(circle_img)
```

---
### 👉 **Figuras preenchidas**

**Para criar uma figura preenchida é necessário indicar um valor negativo como valor do parâmetro *tickness***

```python
# Desenhe um circulo vermelho preenchido no canto inferior direito da imagem
# Mostre o resultado 

draw_img = img_black.copy()

center = (250, 250) 
radius = 60
color = (255, 0, 255) 
thickness = -1 # circulo preenchido
  
circle_img = cv2.circle(draw_img, center=center, radius=radius, color=color, thickness=thickness)

plt.imshow(circle_img)
```

---
### 👉 **line()**

```python
# Desenhe uma linha azul diagonal com espessura de 5 píxels

draw_img = img_black.copy()

pt1 = (400, 100) 
pt2 = (100, 400) 
color = (255, 0, 255) 
thickness = 5

line_img = cv2.line(draw_img, pt1=pt1, pt2=pt2, color=color, thickness=thickness)

plt.imshow(line_img)
```

---
### 👉 **putText()**

**Inserindo texto na imagem**

```python
# insira um texto na imagem utilizando a fonte cv2.FONT_HERSHEY_SIMPLEX

draw_img = img_black.copy()

text = "Thayna"
size = 2
fontFace = cv2.FONT_HERSHEY_SIMPLEX
origin = (100, 100)
color = (255, 0, 255) 
thickness = 5

txt_img = cv2.putText(draw_img, text=text, org=origin, fontFace=fontFace, fontScale=size, color=color, thickness=thickness)

plt.imshow(txt_img)
```
