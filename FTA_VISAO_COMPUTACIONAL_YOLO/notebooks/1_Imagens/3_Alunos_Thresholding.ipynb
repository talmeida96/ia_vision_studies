## **Diferentes tipos de** ***tresholds***

###  **THRESH_BINARY**

**A funo de threshold retorna duas informaes; No caso do THRESH_BINARY:**
- **O primeiro valor indica o prprio valor de threshhold informado na funo como parmetro "thresh"**;
- **A segunda informao  a matriz com a imagem resultado da aplicao do threshold**.

```python
# Aplique o treshold do tipo THRESH_BINARY na imagem em preto e branco e use o valor de thresh = 127
# Mostre o resultado das duas variveis retornadas

img = cv2.imread("../DATA/rainbow.jpg", 0)
ret, thresh1 = cv2.threshold(src=img,thresh=127,maxval=255,type=cv2.THRESH_BINARY)
display(ret)
plt.imshow(thresh1, cmap='gray')
```

---
###  **THRESH_BINARY_INV**

```python
# Aplique o treshold do tipo THRESH_BINARY_INV na imagem em preto e branco e use o valor de thresh = 127
# Mostre o resultado das duas variveis retornadas

img = cv2.imread("../DATA/rainbow.jpg", 0)
ret, thresh1 = cv2.threshold(src=img,thresh=127,maxval=255,type=cv2.THRESH_BINARY_INV)
display(ret)
plt.imshow(thresh1, cmap='gray')
```
