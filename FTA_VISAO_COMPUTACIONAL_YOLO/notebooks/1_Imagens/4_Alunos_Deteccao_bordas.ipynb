### 👉 **Canny()**

```python
# Use a função Canny com os dois parâmetros de threshold iguais a 127 
# e mostre o resultado na tela

img = cv2.imread("../DATA/tesla.jpg")

img_new = cv2.Canny(img, threshold1=127, threshold2=127)

plt.imshow(img_new)
```

```python
# Use a função Canny com os dois parâmetros de threshold diferentes: 0 e 255 
# Mostre o resultado na tela

img = cv2.imread("../DATA/tesla.jpg")

img_new = cv2.Canny(img, threshold1=0, threshold2=255)

plt.imshow(img_new)
```

```python
# Use a função Canny com os dois parâmetros de threshold diferentes: 0 e 10 
# Mostre o resultado na tela

img = cv2.imread("../DATA/tesla.jpg")

img_new = cv2.Canny(img, threshold1=0, threshold2=10)

plt.imshow(img_new)
```

💡 **Note que a diferença entre os valores de threshold influencia em como as arestas são criadas: mais ou menos densas**
